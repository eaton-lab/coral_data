```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)
```

```python
depth = np.arange(1,51)
```

```python
seasonal_cycle = np.sin(2 * np.pi * depth / 12)
```

```python
temp_trend = -0.02 * depth
```

```python
noise = np.random.normal(0, 0.1, size=len(depth))
```

```python
do18 = -3 + seasonal_cycle + temp_trend + noise
```

```python
data = pd.DataFrame({'Depth (mm)':depth, 'δ18O (‰)':do18})
```

```python
data.head()
```

```python
data.to_csv('fake_coral_geochemistry.csv', index=False)
```
