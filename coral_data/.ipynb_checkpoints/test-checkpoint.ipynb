```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
np.random.seed(42) # the integer inside doesn't matter, it is just making sure that the dataset is the same each time it runs
```

```python
depth = np.arange(1,51)
```

```python
seasonal_cycle = np.sin(2 * np.pi * depth / 12)
```

```python
temp_trend = -0.02 * depth
```

```python
noise = np.random.normal(0, 0.1, size=len(depth)) # mall noise, np.random.normal() is a function from NumPy that generates random numbers from a normal (Gaussian) distribution.
```

```python
do18 = -3 + seasonal_cycle + temp_trend + noise
```

```python
do18_data = pd.DataFrame({'Depth (mm)':depth, 'δ18O (‰)':do18})
```

```python
do18_data.head()
```

```python
do18_data.to_csv('fake_coral_geochemistry.csv', index=False)
```

```python
fig_d18o = plt.figure(figsize=(8, 5))
plt.style.use('classic')
plt.plot(depth, do18, marker='o', linestyle='-', color='black', label='Simulated δ18O')
plt.xlabel('Depth (mm)')
plt.ylabel('δ18O (‰)')
plt.title('Simulated Coral δ18O Dataset')
plt.gca().invert_xaxis()  # Depth increases downward in time
plt.grid()
plt.show()
```

```python
# Step 1: Generate a date range (2005–2025) with monthly intervals.

# Again, set random seed for reproducibility
np.random.seed(42)

# Generate a date range for 20 years of monthly data (2005-2025) in date form (FOR FIGURE ONLY)
date_range = pd.date_range(start="2005-01-01", end="2025-12-01", freq='MS')

# # Generate a date range for 20 years of monthly data (2005-2025) in the "years ago" format (FOR THE .TXT FILE)
years_ago = np.linspace(0, 20, num=21 * 12)  # 21 years × 12 months
```

```python
# Generate fake American Samoa SST record (degrees C)
sst_baseline = 28.0  # Average SST in American Samoa
sst_seasonal_variation = 1.0 * np.sin(2 * np.pi * date_range.month / 12)  # Seasonal cycle
sst_trend = 0.02 * (date_range.year - 2005)  # Warming trend (~0.02 degrees C per year)
sst_noise = np.random.normal(0, 0.3, len(date_range))  # Small noise, np.random.normal() is a function from NumPy that generates random numbers from a normal (Gaussian) distribution.
sst = sst_baseline + sst_seasonal_variation + sst_trend + sst_noise # Final sst with the baseline, plus the seasonal variation, plus the warming trend, and normal noise
```

```python
# save sst with year data as dataframe
sst_data = pd.DataFrame({'Years ago':years_ago, 'SST (degree C)':sst})
sst_data.head()
```

```python
sst_data.to_csv('simulated_amsamoa_sst_dataset.csv', index=False)
```

```python
fig_temp = plt.figure(figsize=(10, 5))
plt.style.use('classic')
plt.plot(date_range, sst, marker='o', linestyle='-', color='black', label='Simulated SSTs')
plt.xlabel('Year')
plt.ylabel('SST (degrees C)')
plt.title('Simulated American Samoa SSTs')
plt.gca()  # Depth increases downward in time
plt.ylim(26, 30)
plt.grid()
plt.show()
```
